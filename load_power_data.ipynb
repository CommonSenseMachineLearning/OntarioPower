```
demand_input_data = pd.DataFrame() #create empty DataFrame

#load data files for 2015 to 2020:
for i in range(2015,2021):
  
  #build path and file name
  file_name = f'PUB_DemandZonal_{i}.csv'
  print(i,'file:', file_name)

  #load file into temp DataFrame
  temp_demand = pd.read_csv(input_data_path['demand'] + file_name, skiprows=3)
  print('shape:', temp_demand.shape, '\n')
  
  #add temp file to final DataFrame
  demand_input_data = pd.concat([demand_input_data, temp_demand])

#preview final data
print('Final shape:', demand_input_data.shape)
display(demand_input_data)

print('\nRandom Sample:\n')
display(demand_input_data.sample(5).sort_values(by='Date'))
```
